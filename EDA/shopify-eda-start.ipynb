```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10,10))
sns.histplot(df['total_checkouts'], color='red', kde=True)
```

```python
import plotly.express as px
import plotly.graph_objs as go
from plotly.offline import iplot

px.bar(df, x='referrer_host', y='total_checkouts', title='Total amount of checkouts per referrer')
```

```python
px.bar(df, x='location_country', y='total_checkouts', title='Total amount of checkouts per country')
```

```python
df_sales_per_referrer = df.groupby('referrer_host')['total_checkouts'].mean()
df_sales_per_referrer.plot.bar(figsize=(20,10))
```

```python
df_sales_per_country = df.groupby('location_country')['total_checkouts'].mean()
df_sales_per_country.plot.bar(figsize=(20,10))
```
